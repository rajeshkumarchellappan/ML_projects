## Building Model

```python
X=df1.drop('class',axis=1) #features
y=df1['class'] #target
```

```python
#Splitting dataset for training and testing
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

### K-Nearest Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
```

```python
# Creating instance of KNN
clf = KNN(n_neighbors = 4,weights = 'distance')

# Fitting the model
clf.fit(X_train, y_train)

# Predicting over the Train Set and calculating F1
test_predict = clf.predict(X_test)

k = f1_score(y_test, test_predict,pos_label=2)
print('Test F1 Score    ', k )
print("classification_report for KNN")

print("..."*10)

print(classification_report(y_test, test_predict))
```

### Elbow for Classifier

```python
def Elbow(K):
    #initiating empty list
    test_error = []
   
    #training model for evey value of K
    for i in K:
        #Instance oh KNN
        clf = KNN(n_neighbors = i,weights = 'distance')
        clf.fit(X_train, y_train)
        # Appending F1 scores to empty list claculated using the predictions
        tmp = clf.predict(X_test)
        tmp1 = f1_score(y_test, tmp,pos_label=2)
        error = 1-tmp1
        test_error.append(error)
    
    return test_error
```

```python
#Defining K range
k = range(2, 20, 2)
```

```python
#calling above defined function
test = Elbow(k)
```

```python
# plotting the Curves
plt.plot(k, test)
plt.xlabel('K Neighbors')
plt.ylabel('Test error')
plt.title('Elbow Curve for test');
```

```python
# Creating instance of KNN
clf = KNN(n_neighbors = 4,weights = 'distance')

# Fitting the model
clf.fit(X_train, y_train)

# Predicting over the Train Set and calculating F1
test_predict = clf.predict(X_test)
k = f1_score(test_predict, y_test,pos_label=2)
print('Test F1 Score    ', k )
print("classification_report for KNN")

print("..."*10)

print(classification_report(y_test, test_predict))
```
